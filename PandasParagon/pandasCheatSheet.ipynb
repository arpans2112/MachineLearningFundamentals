#### checking missing values for all columns

````python
# Get missing no of data in all columns
def get_missing_values(df):
    missing_dict = {}
    for column in df.columns:
        ds = df[[column]].isnull().value_counts()
        missing_dict[column] = ds.get((True,))
    return missing_dict

missing_dict = get_missing_values(emp_df_with_missing)
missing_dict
````

### Aggregate function against a row and columns
<a href="#0">Click here to Go to Table of content</a>

```python
    1.  df.sum(axis=0)   # sum of all rows values typically form a column
    2.  df.sum(axis=1)   # sum of all columns values typically form a row
    3.  df.mean(axis=0)  # mean/avg of all rows values typically form a column
    4.  df.mean(axis=1)  # mean/avg of all columns values typically form a row
    5.  df.min(axis=0)  # minimum of all rows values typically form a column
    6.  df.min(axis=1)  # minimum of all columns values typically form a row
    7.  df.max(axis=0)  # max of all rows values typically form a column
    8.  df.max(axis=1)  # max of all columns values typically form a row
    9.  df.std(axis=0)  # standard deviation of all rows values typically form a column
    10. df.std(axis=1)  # standard deviation of all columns values typically form a row
