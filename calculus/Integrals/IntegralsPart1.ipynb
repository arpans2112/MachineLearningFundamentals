# Area Under the Curve Under ROC Curve



When we don't have a function but we do have (x,y) coordinates, we can use the scikit learn library's auc() method, 
which uses a numerical approach (the trapezoidal rule) to find the area under the curve described by the coordinates

```python
from sklearn.metrics import auc
```

from the slides, the (x,y) coordinates of our hot dog detcting ROC Curve are. 

(0,0)
(0,0.5)
(0.5,0.5)
(0.5,1)
(1,1)


```python
xs = [0,0, 0.5, 0.5, 1]
ys = [0,0.5, 0.5, 1, 1]
```

```python
areaUnderCurve = auc(xs,ys)
```
