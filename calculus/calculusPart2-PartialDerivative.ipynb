```python
import os 
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import numpy as np
import matplotlib.pyplot as plt
import torch 
import math
```

```python
# Plot a equation Line

xs = np.linspace(-3, 3, 1000)
# print(x)
# print(xs)

def f(my_x,my_y):
    return my_x**2 - my_y**2

zs_wrt_x = f(xs,0)
# print(zs_wirt_x)


def plotf(x,y):
    fig, ax = plt.subplots()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')
    plt.xlabel('x')
    plt.ylabel('z', rotation=0) 
    ax.plot(x,y)

plotf(xs, zs_wrt_x)

```

```python
def delz_delx(my_x, my_y):
    return 2*my_x


def delz_dely(my_x, my_y):
    return -2*my_y
```

```python
x_samples = [-2, -1, 0 , 1, 2]
```

```python
colors = ['red', 'oranage', 'green', 'blue', 'purpule']
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.ylim(-1,9)

def point_and_tangent_wrt_x(my_xs, my_x, my_y, my_f, fprime, col):
    
#   calculate the value of z , with x_samples points , y = 0
    my_z = my_f(my_x, my_y)
#     print(my_z,  my_x, my_y, my_f, fprime, col)
#     scatter plot my_x, my_z when y = 0
    plt.scatter(my_x, my_z,  zorder=3)
    
    tangent_m = fprime(my_x, my_y)
    tangent_b = my_z - tangent_m*my_x
    tangent_line = tangent_m*my_xs + tangent_b
#     print(tangent_line)
    plt.plot(my_xs, tangent_line,  linestyle='dashed', linewidth=0.7, zorder=3)
    
def point_and_tangent_wrt_y(my_xs, my_x, my_y, my_f, fprime, col):
    
#   calculate the value of z , with x_samples points , y = 0
    my_z = my_f(my_x, my_y)
#     print(my_z,  my_x, my_y, my_f, fprime, col)
#     scatter plot my_x, my_z when y = 0
    plt.scatter(my_y, my_z,  zorder=3)
    
    tangent_m = fprime(my_x, my_y)
    tangent_b = my_z - tangent_m*my_y
    tangent_line = tangent_m*my_xs + tangent_b
#     print(tangent_line)
    plt.plot(my_xs, tangent_line,  linestyle='dashed', linewidth=0.7, zorder=3)
    
for i, x in enumerate(x_samples):
    point_and_tangent_wrt_x(xs,x,0, f, delz_delx, colors[i])
ax.plot(xs, zs_wrt_x)


fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.ylim(-9,1)

y_samples = x_samples.copy()
ys= np.linspace(-3,3, 1000)
zs_wrt_y = f(0, ys)


for i, y in enumerate(y_samples):
    
    point_and_tangent_wrt_y(ys,0,y, f, delz_dely, colors[i])

    ax.plot(ys, zs_wrt_y)
```
