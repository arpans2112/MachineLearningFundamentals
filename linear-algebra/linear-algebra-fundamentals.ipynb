# linear algebra visulatization

### import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

## problem 1 : Draw lines which are intersecting to each other

```python
# create a plot and add labels to the axis
x = np.linspace(-10,10,1000)
y1 = 3*x
y2 = (5*x)/2 + 1

fig, ax = plt.subplots()
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# set limit on x axis and y axis
ax.set_xlim([0,3])
ax.set_ylim([0,8])

# plot line with the cordinates
ax.plot(x,y1,c='green')
ax.plot(x,y2,c='brown')



plt.axvline(x=2,color='purple', linestyle='--')
plt.axhline(y=6, color='purple', linestyle='--')

```

## Problem2: Create parallel Lines

```python
# Below are the parallel Lines. 

y1 = -(9/15)*x + 1
y2 = -(3/5)*x + 2

fig, ax = plt.subplots()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# color = [ "lightbray", "black"]
plt.axvline(x=0, color='black')
plt.axhline(y=0, color='black')

ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.plot(x, y1, c="green")
ax.plot(x, y2, c="brown")
```
