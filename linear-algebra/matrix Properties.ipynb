# Matrix Multiplication

```python
import numpy as np
```

```python
mat_1 = np.array([[3,4],[5,6],[7,8]])
mat_1
```

```python
mat_2 = np.array([[1,9],[2,0]])
mat_2
```

```python
mat_multiplication = np.dot(mat_1, mat_2)
mat_multiplication
```

## Matrix Multiplication is not commuative



```python
mat_a = np.array([[1,9],[2,0]])
mat_b = np.array([[1,9],[2,1]])
mat_multiplication_ab = np.dot(mat_a, mat_b)
print(mat_multiplication_ab)
```

```python
mat_multiplication_ba = np.dot(mat_b, mat_a)
print(mat_multiplication_ba)
```

# Matrix multiplication to another matrix also called applying a matrix to a Vector
can lenearly transform the vector, rotate it, or rescalse it

```python
# vector a
a = np.array([3,1])

#  Identity matrix
I = [[1,0],[0,1]]

resultM = np.dot(I ,a)
```

```python
resultM
```
